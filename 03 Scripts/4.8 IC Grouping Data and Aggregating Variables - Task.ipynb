## Question 4

```python
#create max order
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# create loyalty flag Loyal customer
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# create loyalty flag Regular customer
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# create loyalty flag New customer
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# check output
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

## Question 5

```python
#basic stats by loyalty group
ords_prods_merge.groupby('loyalty_flag').agg({'prices' : ['mean', 'min', 'max']})
```

## Question 6

```python
#create average_spend based on mean purchase price by user_id
ords_prods_merge['avg_spend'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#check output
ords_prods_merge.head(15)
```

```python
# create low spender flag
ords_prods_merge.loc[ords_prods_merge['avg_spend'] < 10, 'spender_flag'] = 'Low spender'
```

```python
# create high spender flag
ords_prods_merge.loc[ords_prods_merge['avg_spend'] > 10, 'spender_flag'] = 'High spender'
```

```python
# check output 
ords_prods_merge[['user_id', 'spender_flag', 'order_number']].head(60)
```

## Question 7

```python
#create order frequency based on median of days_since_last_order
ords_prods_merge['order_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
#check output
ords_prods_merge.head(15)
```

```python
# create non-frequent customer flag
ords_prods_merge.loc[ords_prods_merge['order_frequency'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
# create regular customer flag
ords_prods_merge.loc[(ords_prods_merge['order_frequency'] > 10) & (ords_prods_merge['order_frequency'] <= 20) , 'order_frequency_flag'] = 'Regular customer'
```

```python
# create frequent customer flag
ords_prods_merge.loc[ords_prods_merge['order_frequency'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# check output 
ords_prods_merge[['user_id', 'order_frequency_flag', 'order_frequency']].head(60)
```
